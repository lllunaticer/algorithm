### [2. 两数相加](https://leetcode.cn/problems/add-two-numbers/description/?envType=study-plan-v2&envId=top-interview-150)  
<font color=#FFCC33>中等</font>  
 

简单模拟

```python
from typing import Optional


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        carrier = 0
        root = ListNode()
        cur = root
        while l1 or l2:
            value = (l1.val if l1 else 0) + (l2.val if l2 else 0) + carrier
            cur.next = ListNode(val=value % 10)
            carrier = value // 10
            cur = cur.next
            l1 = l1.next if l1 else None
            l2 = l2.next if l2 else None
        if carrier > 0:
            cur.next = ListNode(val=carrier)
        return root.next


if __name__ == '__main__':
    Solution().addTwoNumbers(
        l1=ListNode(val=2, next=ListNode(val=4, next=ListNode(val=3))),
        l2=ListNode(val=5, next=ListNode(val=6, next=ListNode(val=4)))
    )
```

递归

```python

class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        carrier = 0

        def dfs(l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
            nonlocal carrier
            if not l1 and not l2:
                if carrier > 0:
                    return ListNode(val=carrier)
                else:
                    return None
            value = carrier
            if l1:
                value += l1.val
            if l2:
                value += l2.val
            carrier = value // 10
            return ListNode(val=value % 10, next=dfs(l1.next if l1 else None, l2.next if l2 else None))

        return dfs(l1, l2)


if __name__ == '__main__':
    Solution().addTwoNumbers(
        l1=ListNode(val=2, next=ListNode(val=4, next=ListNode(val=3))),
        l2=ListNode(val=5, next=ListNode(val=6, next=ListNode(val=4)))
    )
```
