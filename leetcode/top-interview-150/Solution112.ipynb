### [112 路径总和](https://leetcode.cn/problems/path-sum/description/?envType=study-plan-v2&envId=top-interview-150)  
<font color=#00FF00>简单</font>  


这道题目，关键是判断何时一条路径走完了的。  
如果用`node is None` 作为是否到达叶子节点的判断，会把下面这种情况下的根节点1也计算为叶子节点(根节点右子树为空)，这是不符合题意的:  
![](../../resource/112.png)  
<font color=#FF3366>错误解法</font>  
下面这种这种解法会fail上面这个用例

```python
class Solution:
    def hasPathSum(self, root: Optional[TreeNode], targetSum: int) -> bool:
        res = False

        def dfs(root: Optional[TreeNode], sum: int) -> None:
            if root is None:
                nonlocal res
                res = res or sum == targetSum
                return

            dfs(root.left, sum + root.val)
            dfs(root.right, sum + root.val)

        dfs(root, 0)
        return res
```

正确的解法应该是在叶子节点做判断:   
`root.left is None and root.right is None` 这个条件保证在  `root`不为空的时候一定是叶子节点，`root`为空单独判断一下就好了

```python
from typing import Optional


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class Solution:
    def hasPathSum(self, root: Optional[TreeNode], targetSum: int) -> bool:
        if root is None:
            return False

        res = False

        def dfs(root: Optional[TreeNode], sum: int) -> None:
            if root is None:
                return

            if root.left is None and root.right is None:
                nonlocal res
                res = res or sum + root.val == targetSum
                return

            dfs(root.left, sum + root.val)
            dfs(root.right, sum + root.val)

        dfs(root, 0)
        return res


if __name__ == '__main__':
    Solution().hasPathSum(root=None, targetSum=0)
```

上面的解法必须要遍历完整个二叉树才会返回答案，下面这个做法可以找到答案就返回结果  
<font color=#00CCFF>如果要在递归中"找到即止"（找到第一个符合条件的就结束），那么通常需要在分支的递归调用前加上return</font>

```python
class Solution:
    def hasPathSum(self, root: Optional[TreeNode], targetSum: int) -> bool:
        if not root:
            return False
        if not root.left and not root.right:
            return targetSum == root.val

        return self.hasPathSum(root.left, targetSum - root.val) or self.hasPathSum(root.right, targetSum - root.val)
```
