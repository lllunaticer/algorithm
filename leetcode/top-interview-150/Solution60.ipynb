### [60 排列序列](https://leetcode.cn/problems/permutation-sequence/description/)  
<font color=#FF3366>困难</font>

阶乘进制（factoradic）选第 k 个排列

```python
import math


class Solution:
    def getPermutation(self, n: int, k: int) -> str:
        f = n
        used = [False] * (n + 1)
        res = ''
        while f > 0:
            s = 0
            ff = math.factorial(f - 1)
            while k > ff:
                k -= ff
                s += 1
            i = 1
            # 找第 s 个未使用的数字
            while i < n + 1:
                if s == 0 and not used[i]:
                    break
                if s > 0 and not used[i]:
                    s -= 1
                i += 1
            res += str(i)
            used[i] = True

            f -= 1
        return res


if __name__ == '__main__':
    print(Solution().getPermutation(4, 1))
                
                
```

优化解法:

```python
import math

class Solution:
    def getPermutation(self, n: int, k: int) -> str:
        nums = [str(i) for i in range(1, n + 1)] # 用pop替代used数组
        k -= 1  # 转成 0-based
        ans = []

        for f in range(n, 0, -1):
            ff = math.factorial(f - 1)
            idx = k // ff
            k %= ff
            ans.append(nums.pop(idx))

        return ''.join(ans)
```
