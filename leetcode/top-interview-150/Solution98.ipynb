### [98 验证二叉搜索树](https://leetcode.cn/problems/validate-binary-search-tree/description/?envType=study-plan-v2&envId=top-interview-150)  
<font color=#FFCC33>中等</font>  

中序遍历到底:

```python
from cmath import inf
from typing import Optional


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class Solution:
    def isValidBST(self, root: Optional[TreeNode]) -> bool:
        before = -inf
        res = True

        def dfs(node: Optional[TreeNode]):
            if node is None:
                return
            dfs(node.left)

            nonlocal res, before
            res = (res and (True if before < node.val else False))
            before = node.val

            dfs(node.right)

        dfs(root)
        return res
```

中序遍历 提前返回:

```python
class Solution:
    def isValidBST(self, root: Optional[TreeNode]) -> bool:
        before = -inf

        def dfs(node: Optional[TreeNode]) -> bool:
            if node is None:
                return True
            left_res = dfs(node.left)  # 遍历左子树
            if not left_res: return False  # 提前返回

            nonlocal before
            if node.val <= before:
                return False
            before = node.val

            return dfs(node.right)

        return dfs(root)
```
