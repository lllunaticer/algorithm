### [138 随机链表的复制](https://leetcode.cn/problems/copy-list-with-random-pointer/description/?envType=study-plan-v2&envId=top-interview-150)  
<font color=#FFCC33>中等</font>

难点在于如何复制random指针，复制 random指针的时候，如何找到原来链表中的random对象，在复制后的链表中对应的random对象。
一个直接的想法是用哈希表保存这个映射关系

```python
# Definition for a Node.
class Node:
    def __init__(self, x: int, next: 'Node' = None, random: 'Node' = None):
        self.val = int(x)
        self.next = next
        self.random = random


class Solution:
    def copyRandomList(self, head: 'Optional[Node]') -> 'Optional[Node]':
        node_dict = {}
        root = Node(x=0)
        copy_cur = root

        # 第一遍复制next链表
        cur = head
        while cur:
            copy = Node(x=cur.val)
            node_dict[cur] = copy

            copy_cur.next = copy

            copy_cur = copy_cur.next
            cur = cur.next

        # 第二遍建立random关系
        copy_cur = root.next
        cur = head
        while cur:
            copy_cur.random = node_dict[cur.random]
            copy_cur = copy_cur.next
            cur = cur.next

        return root.next
```

```python
class Solution:
    def copyRandomList(self, head: 'Optional[Node]') -> 'Optional[Node]':
        if not head:
            return None

        cur = head
        # 构建拼接链表
        while cur:
            copy = Node(x=cur.val)
            copy.next = cur.next
            cur.next = copy
            cur = copy.next

        # 复制random指针
        cur = head
        while cur:
            cur.next.random = cur.random.next if cur.random else None
            cur = cur.next.next

        # 拆开链表
        cur = head
        root = head.next
        new_cur = root

        while cur:
            cur.next = new_cur.next
            new_cur.next = new_cur.next.next if new_cur.next else None
            cur = cur.next
            new_cur = new_cur.next
            
        return root
```
