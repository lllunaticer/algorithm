### [114 二叉树展开为单链表](https://leetcode.cn/problems/flatten-binary-tree-to-linked-list/description/?envType=study-plan-v2&envId=top-interview-150)  
<font color=#FFCC33>中等</font>  
最符合直觉的做法： 考虑按照先序遍历的顺序将二叉树的节点保存下来，然后再依次连接  


```python
from typing import Optional


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class Solution:
    def flatten(self, root: Optional[TreeNode]) -> None:
        """
        Do not return anything, modify root in-place instead.
        """
        if root is None:
            return

        nodes = []

        def dfs(root: Optional[TreeNode]) -> None:
            if root is None:
                return
            nonlocal nodes
            nodes.append(root)
            dfs(root.left)
            dfs(root.right)

        dfs(root)
        for i in range(len(nodes) - 1):
            nodes[i].left = None
            nodes[i].right = nodes[i + 1]
```

```python
class Solution:
    def flatten(self, root: Optional[TreeNode]) -> None:
        """
        Do not return anything, modify root in-place instead.
        """
        head: Optional[TreeNode] = None

        def dfs(root: Optional[TreeNode]) -> None:
            if root is None:
                return
            dfs(root.right)
            dfs(root.left)
            root.left = None
            nonlocal head
            root.right = head
            head = root

        dfs(root)
```

核心思想：后序遍历 + 重构连接
1. 递归函数的职责
这个递归函数不仅展开二叉树，还返回展开后的链表尾节点，这是理解的关键。

2. 代码执行流程

```python
def flatten(self, root):
    if root is None:
        return None
    # 1. 递归展开左右子树
    left_tail = self.flatten(root.left)
    right_tail = self.flatten(root.right)

    # 2. 处理当前节点
    if left_tail:
        left_tail.right = root.right  # 将左子树的尾节点连接到右子树
        root.right = root.left  # 当前节点连接左子树
        root.left = None  # 清空左指针

    # 3. 返回尾节点
    return right_tail or left_tail or root
```

3. 具体示例分析
以树 [1,2,3] 为例：  
    1  
   / \  
  2   3  
执行过程：

节点2：左右子树都为None，返回节点2自身

节点3：左右子树都为None，返回节点3自身

节点1：

left_tail = 节点2

right_tail = 节点3

执行连接操作：

2.right = 3  # 左子树尾节点连接右子树
1.right = 2  # 当前节点连接左子树
1.left = None

返回 right_tail (节点3)
结果： 1 -> 2 -> 3
