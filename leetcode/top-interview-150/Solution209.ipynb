### [209 长度最小的子数组](https://leetcode.cn/problems/minimum-size-subarray-sum/description/?envType=study-plan-v2&envId=top-interview-150)  
<font color=#FFCC33>中等</font>  

单前缀和会超时：

```python
from bisect import bisect_left
from typing import List


class Solution:
    def minSubArrayLen(self, target: int, nums: List[int]) -> int:
        length = len(nums)
        # 1-based
        pre_sum = [0] * (length + 1)
        for i in range(1, length + 1):
            pre_sum[i] = pre_sum[i - 1] + nums[i - 1]

        min_len = 1
        while min_len <= length:
            start = 0
            while start + min_len <= length:
                if pre_sum[start + min_len] - pre_sum[start] >= target:
                    return min_len
                start += 1
            min_len += 1
        return 0
```

注意到前缀和是有序递增的(原数组每个数字都是正的)
因此可以 在前缀和中使用二分法，对于目标的target, 和起始下标i, 找到最小右边界下标j, 使得i到j的前缀和大于等于 target, 从而降低复杂度

```python
class Solution:
    def minSubArrayLen(self, target: int, nums: List[int]) -> int:
        length = len(nums)
        # 1-based
        pre_sum = [0] * (length + 1)
        for i in range(1, length + 1):
            pre_sum[i] = pre_sum[i - 1] + nums[i - 1]

        min_len = inf
        for i in range(length + 1):
            t = target + pre_sum[i]
            bound = bisect_left(pre_sum, t)
            if bound != length+1:
                min_len = min(min_len, bound - i)
        return 0 if min_len == inf else min_len
            
```
