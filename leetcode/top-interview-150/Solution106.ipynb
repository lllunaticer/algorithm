### [106 从中序和后序遍历序列中递归构造二叉树](https://leetcode.cn/problems/construct-binary-tree-from-inorder-and-postorder-traversal/description/?envType=study-plan-v2&envId=top-interview-150)  
<font color=#00CCFF>中等</font>  


这题和 [105](../Solution105.ipynb) 类似， 关键是算对下标

```python
from typing import List, Optional


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class Solution:
    def buildTree(self, inorder: List[int], postorder: List[int]) -> Optional[TreeNode]:
        inorder_dict = {inorder[x]: x for x in range(len(inorder))}

        def dfs(in_left: int, in_right: int, af_left: int, af_right: int) -> Optional[TreeNode]:
            if in_left > in_right or af_left > af_right:
                return None

            value = postorder[af_right]
            p_idx = inorder_dict[value]

            return TreeNode(
                val=value,
                left=dfs(
                    in_left=in_left,
                    in_right=p_idx - 1,
                    af_left=af_left,
                    af_right=af_left + p_idx - in_left - 1
                ),
                right=dfs(
                    in_left=p_idx + 1,
                    in_right=in_right,
                    af_left=af_left + p_idx - in_left,
                    af_right=af_right - 1
                )
            )

        return dfs(in_left=0, in_right=len(inorder) - 1, af_left=0, af_right=len(postorder) - 1)
```
