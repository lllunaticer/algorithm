### [392 判断子序列](https://leetcode.cn/problems/is-subsequence/description/?envType=study-plan-v2&envId=top-interview-150)  
<font color=#00FF00>简单</font>  


```python
class Solution:
    def isSubsequence(self, s: str, t: str) -> bool:
        s_len, t_len = len(s), len(t)
        s_cur, t_cur = 0, 0
        while s_cur < s_len and t_cur < t_len:
            if t[t_cur] == s[s_cur]:
                s_cur += 1
            t_cur += 1
            if s_cur == s_len:
                return True
        return s_cur == s_len
```

大数进阶方案:
上述的解法大量的时间都花在 已知下一个 s中要匹配的字符为x, 从 t中的当前位置往后找到第一个出现x的位置 这件事上。
考虑预处理 t字符串，形成一个快速查找方案，可以快速的找到从 t的某个位置i起，下一个字符 x的位置：
![](../../resource/392.png)

```python
class Solution:
    def isSubsequence(self, s: str, t: str) -> bool:
        # f[i][x] 表示在 t中，从位置i以后，下一个字符x的位置
        s_len, t_len = len(s), len(t)
        f = [[0] * 26 for _ in range(t_len)]

        # 从后往前推算这个数组
        f.append([t_len] * 26)
        for i in range(t_len - 1, -1, -1):
            for x in range(26):
                f[i][x] = i if ord(t[i]) == x + ord('a') else f[i + 1][x]

        cur = 0
        for i in range(s_len):
            if f[cur][ord(t[i]) - ord('a')] == t_len:
                return False
            else:
                cur = f[cur][ord(t[i]) - ord('a')] + 1
        return True
```
