### [206 反转链表](https://leetcode.cn/problems/reverse-linked-list/)  
<font color=#00FF00>简单</font>


头插法:  
![](../../resource/206.png)  

pre 表示前面的节点，cur 表示当前的节点，头插法的意思就是:  
把新节点放到当前的前面

```python
from typing import Optional


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

# 头插法
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        pre = None
        cur = head
        while cur:
            nxt = cur.next
            # 头插法： cur是新节点，cur.next = pre的意思是把cur放到pre前面
            cur.next = pre 
            
            pre = cur
            cur =nxt
        return pre
```
