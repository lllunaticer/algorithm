### [129 求根节点到叶子节点和](https://leetcode.cn/problems/sum-root-to-leaf-numbers/description/?envType=study-plan-v2&envId=top-interview-150)  
<font color=#FFCC33>中等</font>  

这题可以说 [129 求根节点到叶子节点和](https://leetcode.cn/problems/sum-root-to-leaf-numbers/description/?envType=study-plan-v2&envId=top-interview-150) 几乎一样  
核心都是判断什么时候到达叶子节点了，做好路径累计。  
需要有全局变量的版本:

```python
from typing import Optional


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class Solution:
    def sumNumbers(self, root: Optional[TreeNode]) -> int:
        res = 0

        def dfs(root: Optional[TreeNode], path_sum: int) -> None:
            if not root:
                return
            if not root.left and not root.right:
                nonlocal res
                res += path_sum * 10 + root.val
            dfs(root.left, path_sum * 10 + root.val)
            dfs(root.right, path_sum * 10 + root.val)

        dfs(root, 0)
        return res
```

利用DFS返回值的版本, 不需要全局变量

```python
class Solution:
    def sumNumbers(self, root: Optional[TreeNode], path_sum: int = 0) -> int:
        if not root:
            return 0
        path_sum = path_sum * 10 + root.val
        if not root.left and not root.right:
            return path_sum # 到达叶子路径
        # 遍历左右路径加和
        return self.sumNumbers(root.left, path_sum) + self.sumNumbers(root.right, path_sum) 
```
