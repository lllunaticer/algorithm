### [3379 转换数组](https://leetcode.cn/problems/transformed-array/description/?envType=daily-question&envId=2026-02-07)  
<font color=#00FF00>简单</font>  


```python
from typing import List


class Solution:
    def constructTransformedArray(self, nums: List[int]) -> List[int]:
        length = len(nums)
        res = [0] * length
        for index in range(length):
            leftover = nums[index] % length
            step = leftover if leftover > 0 else length + leftover
            real_index = (index + step) % length
            res[index] = nums[real_index]
        return res


if __name__ == '__main__':
    print(Solution().constructTransformedArray(
        nums=[-1,4,-1]
    ))
```
