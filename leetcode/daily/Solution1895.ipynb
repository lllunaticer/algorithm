### [1895 最大的幻方](https://leetcode.cn/problems/largest-magic-square/description/?envType=daily-question&envId=2026-01-18)  
<font color=#FFCC33>中等</font>  

从 0 到可能的最长边 枚举正方形的边长, 枚举正方形左上角顶点位置
注意:
1. 利用前缀和可以快速求行列的和
2. 正方形的边长应该从大到小枚举，遇到符合条件的长度就返回，可以减少计算量

```python
from typing import List


class Solution:
    def largestMagicSquare(self, grid: List[List[int]]) -> int:
        row, column = len(grid), len(grid[0])
        # 正方形可能的最长边是原始矩阵的较短边
        side = min(row, column)
        if side == 1:
            return 1
        # 行前缀和
        row_sum = [[0] * column for _ in range(row)]
        for i in range(row):
            for j in range(column):
                row_sum[i][j] = row_sum[i][j - 1] + grid[i][j] if j > 0 else grid[i][j]

        # 列前缀和
        col_sum = [[0] * column for _ in range(row)]
        for j in range(column):
            for i in range(row):
                col_sum[i][j] = col_sum[i - 1][j] + grid[i][j] if i > 0 else grid[i][j]

        # 枚举正方形边的大小，从可能的最长边长 side 到 2反向枚举，遇到符合条件的就直接返回
        for l in range(side, 1, -1):
            # 从左上角开始枚举
            for i in range(row - l + 1):
                for j in range(column - l + 1):
                    # 第一行
                    first_row_sum = row_sum[i][j + l - 1] - (0 if j == 0 else row_sum[i][j - 1])
                    check = True
                    # 其他行
                    for d in range(1, l):
                        if row_sum[i + d][j + l - 1] - (0 if j == 0 else row_sum[i + d][j - 1]) != first_row_sum:
                            check = False
                            break

                    if not check:
                        continue

                    # 其他列:
                    for d in range(0, l):
                        if col_sum[i + l - 1][j + d] - (0 if i == 0 else col_sum[i - 1][j + d]) != first_row_sum:
                            check = False
                            break

                    if not check:
                        continue

                    # 对角线
                    diag = 0
                    anti_diag = 0
                    for d in range(0, l):
                        diag += grid[i + d][j + d]
                        anti_diag += grid[i + d][j + l - 1 - d]
                    if diag != first_row_sum or anti_diag != first_row_sum:
                        continue
                    return l
        return 1
                    
```
