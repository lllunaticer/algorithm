### [2943 最大化网格图中的正方形空洞面积](https://leetcode.cn/problems/maximize-area-of-square-hole-in-grid/description/?envType=daily-question&envId=2026-01-15)  
<font color=#FFCC33>中等</font>  

读懂题目这题就会了，本质上是求最长连续递增子序列[Solution123](./Solution128.ipynb)。找两个边数组里的最长连续递增序列的长度，其中较短的限制了正方形的最大面积

```python
from typing import List


class Solution:
    def maximizeSquareHoleArea(self, n: int, m: int, hBars: List[int], vBars: List[int]) -> int:
        def find_longest_continual_seq(bars: List[int]) -> int:
            bars.sort()
            res = 1
            tmp = 1
            # 双指针
            left = 0
            right = left + 1
            while right < len(bars):
                if bars[right] == bars[left] + 1:
                    tmp += 1
                    right += 1
                    left += 1
                    res = max(res, tmp)
                else:
                    left = right
                    right = left + 1
                    res = max(res, tmp)
                    tmp = 1
            return res

        length = min(find_longest_continual_seq(hBars), find_longest_continual_seq(vBars))
        return (length + 1) ** 2
```
