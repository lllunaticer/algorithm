对于一个数，它有哪些因子？这不能一眼看出来。  
但反过来，对于一个数 `i`，`i` 的倍数都有因子 `i`。  
枚举 `i=1,2,3,…,10e5`，以及 `i` 的倍数 `j`，把 `j` 的因子个数加一，因子和加上 `i`。  
对于 `nums` 中的因子个数为 `4` 的元素，累加这些元素的因子和，即为答案。  
[链接](https://leetcode.cn/problems/four-divisors/solutions/3867962/mo-ban-yu-chu-li-mei-ge-shu-de-yin-zi-ge-e0v6/)

```python
from typing import List

MAX = 10 ** 5 + 1
divisor_num = [0] * MAX
divisor_sum = [0] * MAX
for i in range(1, MAX):
    for j in range(i, MAX, i):
        divisor_num[j] += 1
        divisor_sum[j] += i


class Solution:
    def sumFourDivisors(self, nums: List[int]) -> int:
        res = 0
        for num in nums:
            if divisor_num[num] == 4:
                res += divisor_sum[num]
        return res
```
