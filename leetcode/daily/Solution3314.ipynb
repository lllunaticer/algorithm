### [3314 构造最小位运算数组I](https://leetcode.cn/problems/construct-the-minimum-bitwise-array-i/description/?envType=daily-question&envId=2026-01-20)  
<font color=#00FF00>简单</font>  

![解法](../../resource/3314.png)

```python
from typing import List


class Solution:
    def minBitwiseArray(self, nums: List[int]) -> List[int]:
        ans = []
        for num in nums:
            res = -1
            bit = 1
            while num & bit != 0:
                res = num - bit
                bit = bit << 1
            ans.append(res)
        return ans
                
                
```
