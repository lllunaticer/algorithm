```C++
// 121
// 212 213 323 312 313 (3 + 2)
// 123
// 212 231 232 312 (2 + 2)

constexpr int mod = 1e9 + 7;

class Solution {
public:
    int numOfWays(int n) {
        size_t preA = 6, preB = 6; // n = 1 时
        while (--n) {
            size_t tmpA = (preA * 3 + preB * 2) % mod;
            size_t tmpB = (preA * 2 + preB * 2) % mod;
            preA = tmpA;
            preB = tmpB;
        }
        return (preA + preB) % mod;
    }
};
```


````python
class Solution:
    def numOfWays(self, n: int) -> int:
        mod = 10**9 + 7  # Python中使用10**9表示1e9
        preA, preB = 6, 6  # n=1时的初始值，Python无需声明变量类型[2](@ref)
        
        # 循环n-1次（C++的while(--n)等效于Python的for循环）
        for _ in range(n - 1):
            tmpA = (preA * 3 + preB * 2) % mod
            tmpB = (preA * 2 + preB * 2) % mod
            preA, preB = tmpA, tmpB  # 并行赋值更新值
        
        return (preA + preB) % mod
    
if __name__ == '__main__':
    Solution().numOfWays(1)
````
