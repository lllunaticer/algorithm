### [1339 分列二叉树的最大乘积](https://leetcode.cn/problems/maximum-product-of-splitted-binary-tree/description/?envType=daily-question&envId=2026-01-07)  
<font color=#FFCC33>中等</font>  

两次`dfs`，
第一遍求二叉树的总和`sum`  
第二遍在dfs的过程中可以拿到每个子树的`sub_sum`，因为题目要求分成两个子树，另一个子树的和可以用`sum - sub_sum`得到。枚举每个子树的总和求乘积的最大值  
一个小优化是可以使用和定积最大定理，避免乘法计算；
另一个优化是可以在第一遍求二叉树总和的过程中，把每个路径上的子树和保存下来，以空间换时间避免第二次的dfs

```python
from typing import Optional


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class Solution:
    def maxProduct(self, root: Optional[TreeNode]) -> int:
        max = 10 ** 9 + 7

        # 第一遍dfs 求二叉树的总和
        def dfs(root: Optional[TreeNode]) -> int:
            if root is None:
                return 0
            return root.val + dfs(root.left) + dfs(root.right)

        sum = dfs(root)
        nearest_sub_sum = 0

        # 第二遍dfs 是在遍历二叉树的过程中，可以拿到每个子树的和
        def dfs2(root: Optional[TreeNode]) -> int:
            if root is None:
                return 0
            sub_sum = root.val + dfs2(root.left) + dfs2(root.right)
            nonlocal nearest_sub_sum
            # 和定积最大定理，a + b = 固定值，a 和 b 的差越小，a * b的值越大
            # sum - a = b , b - a = sum -a -a = sum - 2a
            if abs(sum - nearest_sub_sum * 2) > abs(sum - sub_sum * 2):
                nearest_sub_sum = sub_sum
            return sub_sum

        dfs2(root)

        return ((sum - nearest_sub_sum) * nearest_sub_sum) % max


if __name__ == '__main__':
    print(Solution().maxProduct(root=TreeNode(val=1, left=TreeNode(val=2, left=TreeNode(4), right=TreeNode(val=5)),
                                              right=TreeNode(val=3, left=TreeNode(6)))))
```
