### [1266 访问所有点的最小时间](https://leetcode.cn/problems/minimum-time-visiting-all-points/description/?envType=daily-question&envId=2026-01-12)
<font color=#FFCC33>中等</font>  


```python
from itertools import pairwise
from typing import List


class Solution:
    def minTimeToVisitAllPoints(self, points: List[List[int]]) -> int:
        if len(points) <= 1:
            return 0
        time = 0
        for x, y in pairwise(points):
            dx = abs(y[0] - x[0])
            dy = abs(y[1] - x[1])
            time += dx + dy - min(dx, dy)
        return time
```
