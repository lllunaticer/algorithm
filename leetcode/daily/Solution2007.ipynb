### [2007 从双倍数组中还原数组](https://leetcode.cn/problems/find-original-array-from-doubled-array/description/)  
<font color=#FFCC33>中等</font>  

一般思路

```python
from collections import defaultdict
from typing import List


class Solution:
    def findOriginalArray(self, changed: List[int]) -> List[int]:
        length = len(changed)
        if length % 2 != 0:
            return []
        counter = defaultdict(int)
        for num in changed:
            counter[num] += 1

        changed.sort()
        res = []
        for num in changed:
            if counter[num] == 0:
                continue
            if counter[num * 2] == 0:
                return []
            counter[num * 2] -= 1
            counter[num] -= 1
            res.append(num)
        return res


if __name__ == '__main__':
    Solution().findOriginalArray([2, 1])
```

```python
class Solution:
    def findOriginalArray(self, changed: List[int]) -> List[int]:
        num_counter = {}
        for num in changed:
            num_counter[num] = num_counter.get(num, 0) + 1
            
        zero_cnt = num_counter.pop(0, 0)
        if zero_cnt % 2 != 0:
            return []
        ans = [0] * (zero_cnt // 2)
        for x in num_counter:
            if x % 2 == 0 and x // 2 in num_counter:
                continue
            while x in num_counter:
                x_cnt = num_counter[x]
                if x_cnt > num_counter.get(x * 2,0):
                    return []
                ans.extend([x] * x_cnt)
                if x_cnt < num_counter.get(x * 2,0):
                    num_counter[x * 2] -= x_cnt
                    x *= 2
                else:
                    x *= 4
        return ans


if __name__ == '__main__':
    print(Solution().findOriginalArray([1, 3, 4, 2, 6, 8]))
        
        
```
