### [712, 两个字符串的最小 ASCII删除和](https://leetcode.cn/problems/minimum-ascii-delete-sum-for-two-strings/description/?envType=daily-question&envId=2026-01-10)  
<font color=#FFCC33>中等</font>  


`dp[i][j]` 表示 `s1[0:i]` 和 `s2[0:j]` 的 `最小ASCII删除和` . <font color=#FF3366>这里`s1[0:i]` 表示前i个字符， i是基于下标定义的</font>  
考虑 `s1[i-1]` 和 `s2[j-1]` 这个位置的字符是否相同对结果的影响:
如果  `s1[i-1] == s2[j-1]`:  
在这种情况下，我们选择保留这个位置的字符而不是删除，这在满足字符串相等的条件下，对 `最小ASCII删除和` 这个目标是最有利的，此时:    
    `dp[i][j]` = `dp[i-1][j-1]`
如果 `s1[i-1] != s2[j-1]`:  
在这种情况下，为了使得字符串相等，`s1[i-1]` 和 `s2[j-1]` 必须删除其一，才有可能在后续递归过程中达到字符串相等的可能，此时的是下面两种情况取较小值：  
    `dp[i][j]` = `dp[i-1][j]` + `ascii(s1[i-1])` -- 选择删除 `s1[i-1]`  
    `dp[i][j]` = `dp[i][j-1]` + `ascii(s2[j-1])`  -- 选择删除 `s2[j-1]`  

<font color=#00CCFF>自顶向下的递归写法</font>

i, j 基于下标定义

```python
from functools import cache


class Solution:
    def minimumDeleteSum(self, s1: str, s2: str) -> int:
        @cache
        # i, j 基于下标定义, i, j是下标，从0 开始 (0 based)
        def dfs(i: int, j: int) -> int:
            if i < 0 and j < 0:
                return 0

            # s1为空，需要删除s2的所有剩余字符
            if i < 0:
                return sum(ord(s2[k]) for k in range(j + 1))  # 当s1的下标i<0时, 需要将s2 从0到下标j的位置的字符都统计进来，所以是range(j+1)

            # s2为空，需要删除s1的所有剩余字符
            if j < 0:
                return sum(ord(s1[k]) for k in range(i + 1))

            # 当前字符相同，不需要删除
            if s1[i] == s2[j]:
                return dfs(i=i - 1, j=j - 1)
            else:
                # 字符不同，选择删除代价较小的方案
                return min(dfs(i=i - 1, j=j) + ord(s1[i]), dfs(i=i, j=j - 1) + ord(s2[j]))

        return dfs(len(s1) - 1, len(s2) - 1)


if __name__ == '__main__':
    print(Solution().minimumDeleteSum("delete", "leet"))
```

i, j 基于第几定义

```python
class Solution:
    def minimumDeleteSum(self, s1: str, s2: str) -> int:
        def dfs(i: int, j: int) -> int:
            if i == 0 and j == 0:
                return 0
            if i == 0:
                return sum(ord(s2[k]) for k in range(j))
            if j == 0:
                return sum(ord(s1[k]) for k in range(i))

            if s1[i - 1] == s2[j - 1]:
                return dfs(i - 1, j - 1)
            else:
                return min(dfs(i - 1, j) + ord(s1[i - 1]), dfs(i, j - 1) + ord(s2[j - 1]))

        return dfs(len(s1), len(s2))
```

<font color=#00CCFF>自底向上的动态规划方法</font>

```python
class Solution:
    def minimumDeleteSum(self, s1: str, s2: str) -> int:
        m, n = len(s1), len(s2)
        dp = [[0] * (n + 1) for _ in range(m + 1)]

        for j in range(1, n + 1):
            dp[0][j] = dp[0][j - 1] + ord(s2[j - 1])

        for i in range(1, m + 1):
            dp[i][0] = dp[i - 1][0] + ord(s1[i - 1])

        for i in range(1, m + 1):
            for j in range(1, n + 1):
                if s1[i - 1] == s2[j - 1]:
                    dp[i][j] = dp[i - 1][j - 1]
                else:
                    dp[i][j] = min(dp[i - 1][j] + ord(s1[i - 1]), dp[i][j - 1] + ord(s2[j - 1]))

        return dp[m][n]
```
