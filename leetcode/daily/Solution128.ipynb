### [128 最长连续序列](https://leetcode.cn/problems/longest-consecutive-sequence/description/)  
<font color=#FFCC33>中等</font>

[哈希表O(n)解法](https://leetcode.cn/problems/longest-consecutive-sequence/solutions/3005726/ha-xi-biao-on-zuo-fa-pythonjavacgojsrust-whop/)  
![](../../resource/128.png)

```python
from typing import List


class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        nums_set = set(nums)
        length = len(nums_set)
        ans = 0
        for x in nums_set:
            if x - 1 in nums_set: # 如果 x 不是序列的起点，直接跳过
                continue
            start = x + 1
            while start in nums_set: # 不断查找下一个数是否在哈希集合中
                start += 1
            # 循环结束后，start-1 是最后一个在哈希集合中的数
            ans = max(ans, start- x) # 从 x 到 start-1 一共 y-x 个数
            if ans * 2 >= length:
                break
        return ans

if __name__ == '__main__':
    Solution().longestConsecutive([100, 4, 200, 1, 3, 2])
```
