重要的是，`dfs`函数的思路是什么，怎么划定四个坐标，怎么找到那个可以循环起来的子问题?  
1. 子问题是什么?  是递归的构建左子树和右子树
2. <font color=#FFCC33>子问题的入参是什么?</font> 是左子树和右子树的边界

```python
from typing import List, Optional


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
        inorder_dict = {inorder[x]: x for x in range(len(inorder))}

        def dfs(pre_left: int, pre_right: int, in_left: int, in_right: int) -> Optional[TreeNode]:
            if pre_left > pre_right or in_left > in_right:
                return None
            value = preorder[pre_left]
            p_index = inorder_dict[value]
            return TreeNode(val=value,
                            left=dfs(pre_left=pre_left + 1,
                                     pre_right=p_index - in_left + pre_left,
                                     in_left=in_left,
                                     in_right=p_index - 1
                                     ),
                            right=dfs(pre_left=p_index - in_left + pre_left + 1,
                                      pre_right=pre_right,
                                      in_left=p_index + 1,
                                      in_right=in_right
                                      )
                            )

        return dfs(pre_left=0, pre_right=len(preorder) - 1, in_left=0, in_right=len(inorder) - 1)
        
```
