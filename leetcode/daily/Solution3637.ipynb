### [3637 三段式数组](https://leetcode.cn/problems/trionic-array-i/?envType=daily-question&envId=2026-01-26)  
<font color=#00FF00>简单</font>  

```python
from typing import List


class Solution:
    def isTrionic(self, nums: List[int]) -> bool:
        length = len(nums)
        target_direct = True
        cnt = 1
        pre = 0
        current = 1
        if nums[current] <= nums[pre]:
            return False
        pre += 1
        current += 1
        while current < length:
            if nums[current] == nums[pre]:
                return False
            real_direct = nums[current] > nums[pre]
            if target_direct != real_direct:
                cnt += 1
                if cnt > 3:
                    return False
                target_direct = not target_direct
            pre += 1
            current += 1
        return cnt == 3 and current == length


if __name__ == '__main__':
    print(Solution().isTrionic(nums=[1,3,5,4,2,6]))   
```
