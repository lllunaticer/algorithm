### [1143 最长公共子序列](https://leetcode.cn/problems/longest-common-subsequence/description/)  
<font color=#FFCC33>中等</font>

`LCS(Longest Common Subsequence)` 问题是典型的`DP(动态规划)`问题。  
考虑`dp[i][j]` 表示字符串`text1[0:i]` 和`text2[0:j]` 的 LCS。(`text1[0:i]`表示text1中的前i个字符):  
1. 当 `text1[i-1] == text2[j-1]` 时， `dp[i][j] == dp[i-1][j-1]` 。 这个条件表示当`text1[i-1]==text2[j-1]`时，这个位置上的字符可以作为LCS的一部分，那么`dp[i][j]` 的LCS应该是 `dp[i-1][j-1]`的LCS , 再加上 对这个位置的字符串的选择(+1);
2. 当`text1[i-1] != text2[j-1]`时, 考虑`dp[i][j]`LCS, 则一定是 1. 不选择 `text1[i-1]`, 2. 不选择`text2[j-1]`, 3.都不选 这三种情况之一：  
用转移方程表达就是： `max(dp[i-1][j], dp[i][j-1], dp[i-1][j-1])`   

对于第二种情况，`dp[i-1][j-1]`这种情况的计算 是包含在 `dp[i-1][j]` 和 `dp[i][j-1]` 的过程中的：
```
def dfs(i: int, j:int):
    xxx
    dfs(i-1, j) # 这里
    dfs(i, j-1) # 和这里只调用了 dfs(i-1, j) 和dfs(i, j-1), 没有调用dfs(i-1, j-1)
```
考虑 这样的一个递归过程: 
```
dfs(10, 10)  # i = 10, j = 10
    -> dfs(9, 10)  
        -> dfs(8,10)  
        -> dfs(9,9) # i = 9, j = 9 相对于上面的 i = 10, j = 10, 即 i - 1, j - 1  
    -> dfs(10, 9)  
        -> dfs(9,9)  
        -> dfs(10,8)
```
可以看出, 即使只在函数定义中调用了`dfs(i-1, j)` 和 `dfs(i, j-1)`, 在递归过程中，也最终触及到了 `dfs(i-1, j-1)`。 即`dfs(10,10)` 的调用, 最后递归到了 `dfs(9,9)` 这种情况  

<font color=#FF3366>注意分清楚这题里面i 和 j的 的定义是下标(下标是从 0开始计数) 还是第几个元素(第几是从 1 开始计数)</font>

i 和 j 基于下标定义的 <font color=#00CCFF>自顶向下的递归</font>解法：

````python
from functools import cache


# 基于下标的定义，
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        @cache
        def dfs(i: int, j: int) -> int:
            # 如果任一字符串为空，LCS长度为0
            if i < 0 or j < 0:
                return 0

            if text1[i] == text2[j]:
                return dfs(i=i - 1, j=j - 1) + 1
            else:
                return max(dfs(i=i - 1, j=j), dfs(i=i, j=j - 1))

        return dfs(i=len(text1) - 1, j=len(text2) - 1)
````

i 和 j 基于 第几位 定义的 <font color=#00CCFF>自顶向下的递归解法</font>:

````python
from functools import cache


# i, j基于第几位定义
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        @cache
        def dfs(i: int, j: int) -> int:
            # 如果任一字符串为空，LCS长度为0
            if i == 0 or j == 0:
                return 0

            if text1[i - 1] == text2[j - 1]:
                return dfs(i=i - 1, j=j - 1) + 1
            else:
                return max(dfs(i=i - 1, j=j), dfs(i=i, j=j - 1))

        return dfs(i=len(text1), j=len(text2))
````

自顶向下的递归方法 转 <font color=#00CCFF>自底向上的动态规划方法</font>:

````python
# i, j 表示的前i/j个字符，不是索引下标
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        m, n = len(text1), len(text2)
        dp = [[0] * (n + 1) for _ in range(m + 1)]

        for i in range(1, m + 1):
            for j in range(1, n + 1):
                if text1[i - 1] == text2[j - 1]:
                    dp[i][j] = dp[i - 1][j - 1] + 1
                else:
                    dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
        return dp[m][n]
````
