### [110 平衡二叉树](https://leetcode.cn/problems/balanced-binary-tree/description/?envType=daily-question&envId=2026-02-08)  
<font color=#00FF00>简答</font>  
平衡二叉树 是指该树所有节点的左右子树的高度相差不超过 1。

```python
from typing import Optional


# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class Solution:
    def isBalanced(self, root: Optional[TreeNode]) -> bool:

        def dfs(root: Optional[TreeNode]):
            if not root:
                return 1, True

            left = dfs(root.left)
            if not left[1]:
                return 0, False
            right = dfs(root.right)
            if not right[1]:
                return 0, False
            return max(left[0], right[0]) + 1, abs(left[0] - right[0]) <= 1

        res = dfs(root)
        return res[1]


if __name__ == '__main__':
    print(Solution().isBalanced(
        TreeNode(val=3, left=TreeNode(val=9), right=TreeNode(val=20, left=TreeNode(15), right=TreeNode(7)))))
```
