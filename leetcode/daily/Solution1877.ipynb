### [1877 数组中最大数对和的最小值](https://leetcode.cn/problems/minimize-maximum-pair-sum-in-array/?envType=daily-question&envId=2026-01-22)  
<font color=#FFCC33>中等</font>  


```python
from math import inf
from typing import List


class Solution:
    def minPairSum(self, nums: List[int]) -> int:
        nums.sort()
        res = inf
        length = len(nums)
        for i in range(length // 2):
            res = min(nums[i] + nums[length - i - 1], res)
        return res
```
