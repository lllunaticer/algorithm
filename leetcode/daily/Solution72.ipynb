### [72. 编辑距离](https://leetcode.cn/problems/edit-distance/description/)  
<font color=#FFCC33>中等</font>  

<font color=#00CCFF>自顶向下的递归解法</font>  

定义`dp[i][j]` 表示 `word1` 前`i`个字符，变换到 `word2`中前`j`个字符，最短需要的操作次数  
状态转移:  
1. 增: `dp[i][j] = dp[i][j-1] + 1`: `word1` 的前`i`个字符，变换到`word2`的前 `j`个字符的操作，是在已经完成`word1`的前`i`个字符变换成`word2`的前`j-1`个字符的基础上，在`word2`的第`j`个位置上新增一个字符这个操作达成的(操作+1)  
2. 删: `dp[i][j] = dp[i-1][j] + 1` : `word1` 的前`i`个字符，变换到`word2`的前`j` 个字符的操作，是在已经完成`word1`的前`i-1`个字符变换成`word2`的前`j`个字符的基础上，在`word1`中删除掉`i`个位置上的字符这个操作达成的(操作+1)
3. 改: `dp[i][j] = dp[i-1][j-1] + 1`: `word1` 的前`i`个字符，变换到`word2`的前`j` 个字符的操作, 是在已经完成`word1`的前`i-1`个字符变换成`word2`的前`j-1`个字符的基础上，将`word1`的第`i`个位置上的字符改成`word2`的第`j`个位置上的字符这个操作达成的(操作+1)

```python
from functools import cache


class Solution:
    def minDistance(self, word1: str, word2: str) -> int:

        @cache
        def dfs(i: int, j: int) -> int:
            # 边界条件，word1、word2都为空字符串，则不需要变换
            if i < 0 and j < 0:
                return 0
            # 边界条件, word1为空字符串，则从空字符串 -> word2 是插入word2的所有字符
            if i == 0:
                return j
            # 边界条件, word2为空字符串，则从word1 -> 空字符串 是删除word1的所有字符
            if j == 0:
                return i

            # 处理递推：
            # i 和 j的位置相等的时候，从dp[i-1][j-1]到dp[i][j] 不需要操作
            if word1[i - 1] == word2[j - 1]:
                return dfs(i - 1, j - 1)
            else:
                return min(dfs(i - 1, j) + 1, dfs(i, j - 1) + 1, dfs(i - 1, j - 1) + 1)

        return dfs(len(word1), len(word2))
```

递归转 <font color=#00CCFF>自底向上的动态规划解法</font>

```python
class Solution:
    def minDistance(self, word1: str, word2: str) -> int:
        m, n = len(word1), len(word2)
        dp = [[0] * (n + 1) for _ in range(m + 1)]

        # word1 向 空字符串转移
        for i in range(1, m + 1):
            dp[i][0] = i

        # 空字符串向 word2转移
        for j in range(1, n + 1):
            dp[0][j] = j

        for i in range(1, m + 1):
            for j in range(1, n + 1):
                if word1[i - 1] == word2[j - 1]:
                    dp[i][j] = dp[i - 1][j - 1]
                else:
                    dp[i][j] = min(dp[i - 1][j] + 1, dp[i][j - 1] + 1, dp[i - 1][j - 1] + 1)
        return dp[m][n]
```
