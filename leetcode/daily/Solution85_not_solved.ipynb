### [85 最大矩形](https://leetcode.cn/problems/maximal-rectangle/description/?envType=daily-question&envId=2026-01-11)  
<font color=#FF3366>困难</font>  


```python
from typing import List


class Solution:
    def largetRectangleArea(self, heights: List[int]) -> int:
        length = len(heights)
        if length <= 0:
            return 0
        left, right = [0] * length, [0] * length

        mono_stack = list()
        for i in range(length):
            while mono_stack and heights[mono_stack[-1]] >= heights[i]:
                mono_stack.pop()
            left[i] = mono_stack[-1] if mono_stack else -1

        mono_stack = list()
        for i in range(length - 1, -1, -1):
            while mono_stack and heights[mono_stack[-1]] >= heights[i]:
                mono_stack.pop()
            right[i] = mono_stack[-1] if mono_stack else length

        return max((right[i] - left[i] - 1) * heights[i] for i in range(length))

    def maximalRectangle(self, matrix: List[List[str]]) -> int:
        n = len(matrix[0])
        heights = [0] * (n + 1)
        ans = 0
        for row in matrix:
            for j, c in enumerate(row):
                if c == '0':
                    heights[j] = 0
                else:
                    heights[j] += 1
            ans = max(ans, self.largetRectangleArea(heights))
        return ans
        
```
