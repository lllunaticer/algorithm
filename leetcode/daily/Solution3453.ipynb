### [3454 分割正方形 I](https://leetcode.cn/problems/separate-squares-i/description/?envType=daily-question&envId=2026-01-13)  
<font color=#FFCC33>中等</font>

先计算总面积
然后用二分法枚举y, 使之逼近 1/2 面积

```python
from math import inf
from typing import List


class Solution:
    def separateSquares(self, squares: List[List[int]]) -> float:
        area: float = 0
        low: float = inf
        high: float = -inf
        for s in squares:
            area += s[2] ** 2
            low = min(low, s[1])
            high = max(high, s[1] + s[2])
        while high - low > 1e-5:  # 判断条件用边长比面积更稳定
            bottom_half_area = 0
            mid = (low + high) / 2
            for s in squares:
                if mid <= s[1]:
                    continue
                elif mid <= s[1] + s[2]:
                    bottom_half_area += (mid - s[1]) * s[2]
                else:
                    bottom_half_area += s[2] ** 2

            if bottom_half_area * 2 < area:
                low = mid
            else:
                high = mid
        return low
                
```
