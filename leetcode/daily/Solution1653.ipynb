### [1653 使字符串平衡的最少删除次数](https://leetcode.cn/problems/minimum-deletions-to-make-string-balanced/description/?envType=daily-question&envId=2026-02-07)  
<font color=#FFCC33>中等</font>  

枚举字符串的每个位置，在这个位置之前的所有a 保留，b删除；在这个位置之后的所有b保留，a删除

思路
理解题意： 删除字符，使得 a 全部在b 前面(注意全部为 a 或者全部为 b 是一种符合条件的特殊答案)
解题过程
找一个位置，这个位置之前的 b全部删除，这个位置之后的 a全部删除。这样的位置是可枚举的。对于长度为 n的字符串 s来说连头带尾一共 n + 1个位置;
题意是要找最少删除数，可以利用前缀和统计每个位置之前的 b的数量b_cnt和这个位置之后的a的数量a_cnt，则这个位置符合条件的方案是删除 b_cnt+a_cnt个字符
枚举每个位置，找出最小的 b_cnt+a_cnt

```python
from math import inf


class Solution:
    def minimumDeletions(self, s: str) -> int:
        length = len(s)
        # 分割位置比字符串数量多一个
        a_cnt = [0] * (length + 1)  # 分割线右边的a数量
        b_cnt = [0] * (length + 1)  # 分割线左边的b的数量
        for i in range(length):
            b_cnt[i + 1] = b_cnt[i] + (1 if s[i] == 'b' else 0)
        for i in range(length - 1, -1, -1):
            a_cnt[i - 1] = a_cnt[i] + (1 if s[i] == 'a' else 0)

        min_value = inf
        for i in range(length + 1):
            min_value = min(min_value, b_cnt[i] + a_cnt[i])
        return min_value


if __name__ == '__main__':
    print(Solution().minimumDeletions("aababbab"))
```

考虑 `s` 的最后一个字母:  
+ 如果它是 `b`, 则无需删除, 问题规模缩小, 变成 使 `s` 的前 `n-1` 个字母平衡的最小删除次数
+ 如果它是 `a`:
    + 删除它, 则答案为使 `s`的前`n-1`个字母平衡的最小删除次数加上 1
    + 保留它，那么前面的所有`b`都要删除

设`cntB`为前前`i`个字母中`b`的个数，定义`f[i]` 表示使`s`的前`i`个字母平衡的最少删除次数:
+ 如果第`i`个字母是`b`, 则`f[i] = f[i-1]`;
+ 如果第`i`个字母是`a`, 则`f[i] = min(f[i-1], cntB)`

```python
class Solution:
    def minimumDeletions(self, s: str) -> int:
        f = cnt_b = 0
        for c in s:
            if c == 'b':
                cnt_b += 1  # f不变, f = 上次的f , 即 f[i] = f[i-1]
            else:
                f = min(f + 1, cnt_b)
        return f
```
