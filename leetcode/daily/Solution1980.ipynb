### [1984 学生分数的最小差值](https://leetcode.cn/problems/minimum-difference-between-highest-and-lowest-of-k-scores/description/?envType=daily-question&envId=2026-01-25)  
<font color=#00FF00>简单</font>

```python
from math import inf
from typing import List


class Solution:
    def minimumDifference(self, nums: List[int], k: int) -> int:
        nums.sort()
        length = len(nums)
        min_diff = inf
        k_d = 0
        while k_d + k <= length:
            min_diff = min(min_diff, nums[k + k_d - 1] - nums[k_d])
            k_d += 1
        return min_diff

if __name__ == '__main__':
    print(Solution().minimumDifference([9,4,1,7], 2))
```
