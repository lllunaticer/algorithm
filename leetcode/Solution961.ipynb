### [961 在长度 2N 的数组中找出重复 N 次的元素](https://leetcode.cn/problems/n-repeated-element-in-size-2n-array/description/?envType=daily-question&envId=2025-12-31)

<font color=#00FF00>简单</font>

方法1: 暴力，根据题意，重复超过一次的数就是重复了n次的数

```python
from random import randrange
from typing import List


class Solution:
    def repeatedNTimes(self, nums: List[int]) -> int:
        record = set()
        for num in nums:
            if num in record:
                return num
            else:
                record.add(num)
```

方法2: <font color=#00CCFF>摩尔投票</font>

摩尔投票可以用来解决众数问题，本题中的目标数不是众数  
-- 差一点，众数是在2n个数中，占比超过一半的数, 也就是大于等于 n + 1个；本题的目标数是 n个。
所以可以这样考虑，从这2n个数选一个数特殊考虑，比如选第一个，如果这个数在后续出现过，那它就是目标数；  
否则排除掉它，那么目标数在剩余的数中就是众数，可以用摩尔投票的模版:

为了让出现 n 次的那个数变成绝对众数，我们可以分类讨论：
如果 `nums[0]` 在下标 `[1,2n−1]` 中出现过，那么返回 `nums[0]`。
否则，去掉 `nums[0]`，剩下 `2n−1` 个数，出现次数为 `n` 的那个数变成绝对众数，可以用 169 题的算法解决。

```python
from typing import List


class Solution:
    def repeatedNTimes(self, nums: List[int]) -> int:
        ans, hp = nums[0], 0
        for x in nums[1:]:
            if x == nums[0]:
                return x
            if hp == 0:
                ans, hp = x, 1
            elif hp > 1:
                return ans
            else:
                hp = hp + 1 if ans == x else hp - 1
        return ans


if __name__ == '__main__':
    print(Solution().repeatedNTimes([8, 3, 2, 3]))
```

方法3: <font color=#00CCFF>数学推导</font>  
检查邻近元素  
设出现次数为 `n` 的那个数为 `x`:

如果相邻两个 `x` 之间都至少有一个数，那么 `nums` 至少要有 `2n−1` 个数, 现在我们有 `2n`个数, `2n >= 2n-1` 合法  
如果相邻两个 `x` 之间都至少有两个数，那么 `nums` 至少要有 `3n−2` 个数, 现在我们有 `2n` 个数, `2n >= 3n -2`, 这个条件仅在 `n<=2` 时成立;  
如果相邻两个 `x` 之间都至少有三个数，那么 `nums` 至少要有 `4n-3` 个数, 现在我们有 `2n`个数, `2n >= 4n -3`, 这个条件在 `n <= 1`时成立, 
这对本题的条件来说不成立，因此只要检测 `nums[i]` 和 `nums[i+1]`, `nums[i+2]`, `nums[i+3]`是否相等，

举例:  
如果 `n=2`，这是合法的，例如 `[3,1,2,3]`。  
如果 `n≥3`，那么 `3n−2>2n`，不合法。这意味着，当 `n≥3` 时，一定存在 `nums[i]=nums[i−1]` 或者 `nums[i]=nums[i−2]`。


```python
class Solution:
    def repeatedNTimes(self, nums: List[int]) -> int:
        length = len(nums)
        for i in range(0, length - 1):
            if (i + 1 < length and nums[i + 1] == nums[i]) or (i + 2 < length and nums[i + 2] == nums[i]) or (
                    i + 3 < length and nums[i + 3] == nums[i]):
                return nums[i]
```

方法3: <font color=#00CCFF>随机采样</font>  
由于在2n个数中，某个数占一半。因此每次随机采样两个数，这两个数如果相等，则这个数就是目标数. 用概率论可以证明，这个算法的效率较高

```python
class Solution:
    def repeatedNTimes(self, nums: List[int]) -> int:
        length = len(nums)
        while True:
            i = randrange(length)
            j = randrange(length)
            while j == i:
                j = randrange(length)
            if nums[i] == nums[j]:
                return nums[i]
            
```
