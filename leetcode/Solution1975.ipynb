### [1975 最大方阵和](https://leetcode.cn/problems/maximum-matrix-sum/?envType=daily-question&envId=2026-01-05)  
<font color=#FFCC33>中等</font>  
偶数个负数可以通过两两相消变成0个，奇数个负数可以通过两两相消，只留下最小的那个负数

```python
from math import inf
from typing import List


class Solution:
    def maxMatrixSum(self, matrix: List[List[int]]) -> int:
        smallest = inf
        sum_res = 0
        negative_count = 0
        for line in matrix:
            for num in line:
                if num < 0:
                    negative_count += 1
                if abs(num) < smallest:
                    smallest = abs(num)
                sum_res += abs(num)
        if negative_count % 2 == 0:
            return sum_res
        else:
            return sum_res - smallest * 2
        
        
if __name__ == '__main__':
    print(4%2)
```
