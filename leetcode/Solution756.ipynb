### [756 金字塔转换矩阵](https://leetcode.cn/problems/pyramid-transition-matrix/description/?envType=daily-question&envId=2025-12-29)
<font color=#00CCFF>中等</font>
DFS 遍历


```python
from collections import defaultdict
from typing import List


class Solution:
    def pyramidTransition(self, bottom: str, allowed: List[str]) -> bool:

        allowed_dict = defaultdict(list)
        for element in allowed:
            allowed_dict[element[0:2]].append(element[2])

        # 递归的构建每一层
        def dfs(current_layer: str, next_layer: str, next_layer_idx: int) -> bool:
            if len(current_layer) == 1:
                return True

            if len(next_layer) == len(current_layer) - 1:
                return dfs(current_layer=next_layer, next_layer='', next_layer_idx=0)

            possibles = allowed_dict[current_layer[next_layer_idx:next_layer_idx + 2]]
            if not possibles:
                return False

            for possible in possibles:
                if (dfs(current_layer=current_layer, next_layer=next_layer + possible,
                        next_layer_idx=next_layer_idx + 1)): return True
            return False

        return dfs(current_layer=bottom, next_layer='', next_layer_idx=0)


if __name__ == '__main__':
    print(Solution().pyramidTransition('BCD', ["BCC", "CDE", "CEA", "FFF"]))
```
